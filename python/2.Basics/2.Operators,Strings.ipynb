## Section 1: Operators and Expressions

Operators are special symbols that designate that some sort of computation should be performed. The values that an operator acts on are called **operands**. A sequence of operands and operators is called an **expression**.

### 1.1 Arithmetic Operators

| Operator | Example  | Meaning            | Important Note                                               |
| -------- | -------- | ------------------ | ------------------------------------------------------------ |
| `+`      | `a + b`  | Addition           |                                                              |
| `-`      | `a - b`  | Subtraction        |                                                              |
| `*`      | `a * b`  | Multiplication     |                                                              |
| `/`      | `a / b`  | Division           | **Always returns a `float`** (even `10 / 5` → `2.0`)         |
| `//`     | `a // b` | Floor Division     | Truncates the fractional part (rounds down). `10 // 4` → `2` |
| `%`      | `a % b`  | Modulo (Remainder) | `10 % 4` → `2`                                               |
| `**`     | `a ** b` | Exponentiation     | `2 ** 3` → `8`                                               |

**Examples:**

```python
a = 4
b = 3
print(a / b)   # 1.3333333333333333
print(a // b)  # 1
print(a % b)   # 1
print(a ** b)  # 64
```

### 1.2 Comparison Operators

These operators return a Boolean value (`True` or `False`).

|Operator|Meaning|
|---|---|
|`==`|Equal to|
|`!=`|Not equal to|
|`<`|Less than|
|`<=`|Less than or equal to|
|`>`|Greater than|
|`>=`|Greater than or equal to|

**⚠️ Critical Warning: Comparing Floating-Point Values**  
Due to the limited precision of floating-point representation, **never** compare floats using `==` for exact equality.

```python
# Don't do this!
x = 1.1 + 2.2
print(x == 3.3)   # False (Unbelievable, but true!)
# Correct method: Use a Tolerance
tolerance = 0.00001
print(abs(x - 3.3) < tolerance)  # True
```

### 1.3 Logical Operators

Used to combine Boolean expressions.

|Operator|Example|Meaning|
|---|---|---|
|`not`|`not x`|Reverses the logical sense (if `True`, becomes `False`)|
|`and`|`x and y`|Both must be `True`|
|`or`|`x or y`|At least one must be `True`|

**The Concept of Truthy and Falsy (Evaluation in Boolean Context):**  
Non-Boolean objects can also be evaluated as "true" or "false" in conditional contexts:

- **Falsy:** `0`, `0.0`, `""` (empty string), `[]` (empty list), `None`.
    
- **Truthy:** Everything else.
    

```python
print(bool(0))      # False
print(bool(" "))    # True (A space is also a character)
```

### 1.4 Identity Operators

Check whether two variables refer to the **exact same object** in memory (not just having equal values).

|Operator|Meaning|
|---|---|
|`is`|Do both refer to the same object?|
|`is not`|Do they refer to different objects?|


```python
x = [1, 2, 3]
y = [1, 2, 3]
print(x == y)   # True (Content is equal)
print(x is y)   # False (Two separate lists were created in memory)
y = x
print(x is y)   # True (Now both refer to the same list)
```

### 1.5 Augmented Assignment Operators

A shorthand way to update a variable.

|Shorthand Expression|Full Equivalent|
|---|---|
|`x += 5`|`x = x + 5`|
|`x -= 5`|`x = x - 5`|
|`x *= 2`|`x = x * 2`|
|`x /= 2`|`x = x / 2`|

## Section 2: Advanced String Manipulation

### 2.1 String Operators

- **Concatenation (`+`):** Glues strings together.
    
- **Repetition (`*`):** Repeats a string a specified number of times.
    
- **Membership (`in`):** Checks if a substring exists within another string.
    

```python
print("Hello" + " " + "World")   # Hello World
print("Ha" * 3)                  # HaHaHa
print("fox" in "The quick brown fox")  # True
```

### 2.2 Built-in Functions for Strings

|Function|Use|Example|
|---|---|---|
|`len(s)`|Length of string|`len("Python")` → `6`|
|`str(obj)`|Convert object to string|`str(3.14)` → `"3.14"`|
|`ord(c)`|Unicode code point of character|`ord('A')` → `65`|
|`chr(n)`|Character from Unicode code point|`chr(8364)` → `'€'`|

### 2.3 String Indexing and Slicing

Accessing individual characters or portions of a string by their position.

- **Indexing:** Starts from zero. `s[0]` is the first character. `s[-1]` is the last character.
    
- **Slicing:** Extracting a substring. `s[start:stop:step]`
    
    - `start`: The starting index (included).
        
    - `stop`: The ending index (**not** included).
        
    - `step`: The stride (optional).
        

**Slicing Examples:**

```python
s = "foobar"
print(s[2:5])    # "oba"  (from index 2 up to, but not including, 5)
print(s[:4])     # "foob" (from the beginning up to index 4)
print(s[2:])     # "obar" (from index 2 to the end)
print(s[::-1])   # "raboof" (reversing the string)
print(s[::2])    # "foa" (every other character)
```

### 2.4 Variable Interpolation with f-strings (Modern Formatting)

Since Python 3.6, this is the best way to embed variables within strings.

```python
name = "Alice"
age = 30
# Place an 'f' before the opening quote and put the variable in {}
print(f"My name is {name} and I am {age} years old.")
# Output: My name is Alice and I am 30 years old.
```

### 2.5 String Immutability

You **cannot** directly change a character in a string. Strings in Python are immutable.

```python
s = "Python"
# s[0] = "J"  # Error! TypeError
# Solution: Build a new string
s = "J" + s[1:]
print(s)  # Jython
```

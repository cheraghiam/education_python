### 1. Introduction to Iteration

**Iteration** means executing the same block of code repeatedly. There are two types:

|Type|Description|Python Keyword|
|---|---|---|
|**Indefinite**|Repeats as long as a condition is true|`while`|
|**Definite**|Repeats a specified number of times (over a collection)|`for`|

---


### 2. The `while` Loop (Indefinite Iteration)

```text
while <expr>:
    <statement(s)>
```

- `<expr>` is evaluated in Boolean context **before** each iteration.
    
- If truthy → loop body executes; if falsy → loop terminates.
    
- The controlling variable must be updated **inside** the loop body.
    

````python
n = 5
while n > 0:
    n -= 1
    print(n)
# Output: 4, 3, 2, 1, 0
````

**Loop over a list using falsiness of empty list:**

````python
a = ['foo', 'bar', 'baz']
while a:
    print(a.pop(-1))
````


---
### 3. The `break` and `continue` Statements

|Statement|Action|
|---|---|
|`break`|Terminates the **entire loop** immediately|
|`continue`|Skips the **current iteration** and goes to the next|

**`break`** example:

````python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        break
    print(n)
print('Loop ended.')
# Output: 4, 3, Loop ended.
````

**`continue` example:**

````python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        continue
    print(n)
print('Loop ended.')
````


---
### 4. The `else` Clause with `while

```text
while <expr>:
    <statement(s)>
else:
    <additional_statement(s)>
```

- The `else` block executes **only if** the loop terminates normally (condition becomes false).
    
- The `else` block **does NOT execute** if the loop is exited via `break`.
    

````python
# With normal termination
n = 5
while n > 0:
    n -= 1
    print(n)
else:
    print('Loop done.')
# Output: 4, 3, 2, 1, 0, Loop done.
# With break
n = 5
while n > 0:
    n -= 1
    print(n)
    if n == 2:
        break
else:
    print('Loop done.')
# Output: 4, 3, 2  (else block skipped)
````

**Practical use:** Searching a list

````python
a = ['foo', 'bar', 'baz', 'qux']
s = 'Corge'
i = 0
while i < len(a):
    if a[i] == s:
        break
    i += 1
else:
    print(s, 'not found in list.')
# Output: Corge not found in list.
````


> **Note:** In practice, use `in` operator or `.index()` method for searching, not a `while` loop.

### 5. Infinite Loops



```text
while True:
    # Do something
    if <condition>:
        break
```

- `while True` runs forever unless a `break` is encountered.
    
- Press `Ctrl+C` to force-stop an infinite loop.
    
- Common pattern for services, servers, or loops with multiple exit conditions.
    

```text
while True:
    if <expr1>:    # Termination condition 1
        break
    if <expr2>:    # Termination condition 2
        break

```

### 6. Nested `while` Loops

Loops can be nested inside each other, and mixed with `if` statements:

````python
a = ['foo', 'bar']
while len(a):
    print(a[0])
    del a[0]
    
    b = ['baz', 'qux']
    while len(b):
        print('>', b[0])
        del b[0]
# Output: foo, > baz, > qux, bar, > baz, > qux
````

- `break`/`continue` apply to the **nearest enclosing loop**.

## 8. The `for` Loop (Definite Iteration) – Expanded Section

### 8.1 Basic Syntax

Python's `for` loop is a **collection-based** loop. Unlike C or Java which use numeric loops, Python iterates directly over elements:

```text
for <var> in <iterable>:
    <statement(s)>

- `<var>` takes the value of each element in `<iterable>` one at a time.
```

    
- The loop continues until all elements have been processed.
    

````python
a = ['foo', 'bar', 'baz']
for i in a:
    print(i)
````

### 8.2 Types of `for` Loops in Programming (Background)

To understand Python's design, it's useful to know the three main types of `for` loops in programming:

|Type|Example Syntax|Used In|
|---|---|---|
|**Numeric Range**|`for i = 1 to 10`|BASIC, Algol, Pascal|
|**Three-Expression**|`for (i=1; i<=10; i++)`|C, C++, Java, PHP, Perl|
|**Collection-Based**|`for i in collection`|**Python**, Perl (`foreach`), PHP (`foreach`)|

Python **only** implements the collection-based approach, which proves to be extremely versatile.

### 8.3 Iterables – What Can You Loop Over?

An **iterable** is any object that can be passed to `iter()` and used in a `for` loop.

**Iterable types in Python:**

|Type|Example|Iterator Type|
|---|---|---|
|String|`'foobar'`|`str_iterator`|
|List|`['foo', 'bar', 'baz']`|`list_iterator`|
|Tuple|`('foo', 'bar', 'baz')`|`tuple_iterator`|
|Set|`{'foo', 'bar', 'baz'}`|`set_iterator`|
|Dictionary (keys)|`{'foo': 1, 'bar': 2}`|`dict_keyiterator`|
|Range|`range(5)`|`range_iterator`|
|File objects|`open('file.txt')`|File iterator|

**Non-iterable types:** `int`, `float`, `bool`, `None`, functions.


````python
iter(42)       # TypeError: 'int' object is not iterable
iter(3.14)     # TypeError: 'float' object is not iterable
iter(len)      # TypeError: 'builtin_function_or_method' object is not iterable
````

### 8.4 Iterators – The Engine Behind `for` Loops

An **iterator** is a value producer that yields successive values from an associated iterable.

|Term|Meaning|Key Function|
|---|---|---|
|**Iterable**|An object that can be iterated over|`iter(obj)`|
|**Iterator**|An object that produces successive values|`next(iterator)`|



````python
a = ['foo', 'bar', 'baz']
itr = iter(a)       # Get iterator
next(itr)           # 'foo'
next(itr)           # 'bar'
next(itr)           # 'baz'
next(itr)           # Raises StopIteration exception
````

**How the `for` loop actually works internally:**

````python
for i in a:
    print(a)
````

Is equivalent to:

````python
itr = iter(a)
while True:
    try:
        i = next(itr)
        print(i)
    except StopIteration:
        break
````

1. Python calls `iter(a)` to get an iterator.
    
2. Calls `next()` repeatedly inside a try-except block.
    
3. When `StopIteration` is raised, the loop terminates.
    

**Key properties of iterators:**

- They are **stateful** – they remember which value to return next.
    
- They can only go **forward** – you cannot go backward (`prev()` does not exist).
    
- Multiple iterators on the same object are **independent**:
    


````python
a = ['foo', 'bar', 'baz']
itr1 = iter(a)
itr2 = iter(a)
next(itr1)  # 'foo'
next(itr1)  # 'bar'
next(itr2)  # 'foo'  (itr2 is still at the beginning!)
````

- You can grab all remaining values at once with `list()`, `tuple()`, or `set()`:

````python
itr = iter(a)
list(itr)   # ['foo', 'bar', 'baz']
````

> **Caution:** This forces all values to be generated at once. For large datasets, this can consume significant time and memory. Iterators are **"lazy"** by nature – values are generated only when requested.

### 8.5 Iterating Through a Dictionary (Expanded

````python
d = {'foo': 1, 'bar': 2, 'baz': 3}
````


**Iterating over keys (default):**



````python
for k in d:
    print(k)                    # foo, bar, baz
````


**Iterating over values:**

                

````python
for v in d.values():
    print(v)                    # 1, 2, 3
````

**Iterating over key-value pairs (Pythonic way):**

````python
for k, v in d.items():
    print(f'k={k}, v={v}')
# k=foo, v=1
# k=bar, v=2
# k=baz, v=3
````

This works because `d.items()` returns an iterable of tuples, and unpacking assigns each tuple to `k, v`:

````python
# Understanding tuple unpacking in for loops
for i, j in [(1, 2), (3, 4), (5, 6)]:
    print(i, j)
# Output: 1 2, 3 4, 5 6

### 8.6 The `range()` Function (Expanded)
````

Python's way of creating numeric sequences efficiently.

|Form|Description|Example|
|---|---|---|
|`range(end)`|0 to `end-1`|`list(range(5))` → `[0,1,2,3,4]`|
|`range(start, end)`|`start` to `end-1`|`list(range(2,6))` → `[2,3,4,5]`|
|`range(start, end, step)`|With step interval|`list(range(5,20,3))` → `[5,8,11,14,17]`|


````python
#Positive step
list(range(5, 10))       # [5, 6, 7, 8, 9]
list(range(5, 10, 1))    # [5, 6, 7, 8, 9]  (same)
# Negative step (counting down)
list(range(5, -5, -1))   # [5, 4, 3, 2, 1, 0, -1, -2, -3, -4]
# Empty range (start ≥ end with positive step, or start ≤ end with negative step)
list(range(5, -5))       # [] (needs negative step)
list(range(10, 5))       # [] (needs negative step)
````

Important characteristics:**

- `range` is **lazy** – values are generated on demand, not stored all at once.
    
- `range` objects are immutable sequence types, not actual lists.
    
- All parameters must be integers (can be negative).
    

### 8.7 `break`, `continue`, and `else` in `for` Loops

These work identically to their `while` loop counterparts.

**`break`** – exits the entire loop:

````python
for i in ['foo', 'bar', 'baz', 'qux']:
    if 'b' in i:
        break
    print(i)
````

**`continue`** – skips to next iteration:

````python
for i in ['foo', 'bar', 'baz', 'qux']:
    if 'b' in i:
        continue
    print(i)
````

**`else`** – runs only if loop completes without `break`:

````python
#Else executes
for i in ['foo', 'bar', 'baz']:
    print(i)
else:
    print('Done.')
# Output: foo, bar, baz, Done.
# Else skipped
for i in ['foo', 'bar', 'baz']:
    if i == 'bar':
        break
    print(i)
else:
    print('Done.')
# Output: foo  (else not executed)
````


**Practical use of `else`:** checking if an item was found without needing a flag variable.

````python
#Without else (using a flag)
found = False
for item in data:
    if item == target:
        found = True
        break
if not found:
    print("Not found")
# With else (cleaner)
for item in data:
    if item == target:
        break
else:
    print("Not found")


# You Can do it ........
````

### 9. Comparing `while` and `for` Loops

|Feature|`while`|`for`|
|---|---|---|
|Use case|Unknown number of iterations|Known collection or range|
|Condition|Repeats while condition is truthy|Iterates over each element|
|Risk|Infinite loop if condition never becomes false|No infinite loop (finite collection)|
|Empty case|Body may never execute|Body may never execute (empty iterable)|
|Pythonic style|Less common for iteration|Preferred for iterating over collections|

**When to use which:**

- Use `for` when iterating over a known collection (list, dict, range, etc.).
    
- Use `while` when the number of iterations depends on a changing condition (user input, network, file reading, etc.).
    

---


### 10. Summary

|Concept|Key Points|
|---|---|
|`while`|Repeats while condition is truthy. Checked at top of loop.|
|`for`|Iterates over each item in an iterable (list, dict, range, etc.).|
|Iterable|Any object that can be passed to `iter()` (strings, lists, dicts, etc.).|
|Iterator|Produces values via `next()`. Maintains internal state.|
|`iter()` / `next()`|`iter(obj)` creates iterator; `next(itr)` returns next value.|
|`StopIteration`|Exception raised when iterator is exhausted.|
|`break`|Exits the entire loop immediately.|
|`continue`|Skips current iteration, continues with next.|
|`else` on loops|Runs after loop **only if** loop finishes normally (no `break`).|
|`range(start, end, step)`|Generates numeric sequences lazily.|
|`dict.items()`|Iterates key-value pairs in a dictionary with unpacking.|
|Infinite Loop|`while True` – must contain a `break` to avoid running forever.|
|Nested Loops|Loops inside loops. `break`/`continue` affect nearest enclosing loop.|

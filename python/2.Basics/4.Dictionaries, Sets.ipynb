## Section 1: Dictionaries

Python provides another composite data type called a **dictionary**, which is similar to a list in that it is a collection of objects.

### Shared Characteristics of Dictionaries and Lists:

- **Both are mutable.**
    
- **Both are dynamic.** They can grow and shrink as needed.
    
- **Both can be nested.** A list can contain another list. A dictionary can contain another dictionary. A dictionary can also contain a list, and vice versa.
    

### Key Difference Between Dictionaries and Lists:

- **List elements** are accessed by their position in the list, via **indexing**.
    
- **Dictionary elements** are accessed via **keys**.
    

---


1.1 Defining a Dictionary

Dictionaries are Python's implementation of a data structure more generally known as an **associative array**. A dictionary consists of a collection of **key-value pairs**. Each key-value pair maps the key to its associated value.

**Method 1: Using Curly Braces `{}`**

You can define a dictionary by enclosing a comma-separated list of key-value pairs in **curly braces (`{}`)** . A colon (`:`) separates each key from its associated value:


```text
d = {
    <key>: <value>,
    <key>: <value>,
      .
      .
      .
    <key>: <value>
}
```

**Example:**

````python
MLB_team = {
    'Colorado' : 'Rockies',
    'Boston'   : 'Red Sox',
    'Minnesota': 'Twins',
    'Milwaukee': 'Brewers',
    'Seattle'  : 'Mariners'
}
````

Method 2: Using the `dict()` Function**

You can also construct a dictionary with the built-in `dict()` function. The argument to `dict()` should be a sequence of key-value pairs. A list of tuples works well for this:

```text
d = dict([
    (<key>, <value>),
    (<key>, <value>),
      .
      .
      .
    (<key>, <value>)
])
```

**Example:**

````python
MLB_team = dict([
    ('Colorado', 'Rockies'),
    ('Boston', 'Red Sox'),
    ('Minnesota', 'Twins'),
    ('Milwaukee', 'Brewers'),
    ('Seattle', 'Mariners')
])
````

Method 3: Using Keyword Arguments**

If the key values are simple strings, they can be specified as keyword arguments. This is yet another way to define `MLB_team`:

````python
MLB_team = dict(
    Colorado='Rockies',
    Boston='Red Sox',
    Minnesota='Twins',
    Milwaukee='Brewers',
    Seattle='Mariners'
)
````

Once defined, you can display a dictionary's contents just like a list:

````python
type(MLB_team)  # <class 'dict'>
MLB_team
# Output:
# {'Colorado': 'Rockies',
#  'Boston': 'Red Sox',
#  'Minnesota': 'Twins',
#  'Milwaukee': 'Brewers',
#  'Seattle': 'Mariners'}
````

> **Important Note:** The entries in the dictionary display in the order they were defined. However, this order is irrelevant when it comes to retrieving elements. **Dictionary elements are not accessed by numerical index:

````python
 MLB_team[1]  # KeyError: 1
````

### 1.3 Dictionary Keys vs. List Indices

The interpreter raises the same exception, `KeyError`, when a dictionary is accessed with either an undefined key or by a numeric index. In fact, it's the same error. In the latter case, `[1]` looks like a numerical index, but it isn't.

You **can use integers as dictionary keys**:

````python
d = {0: 'a', 1: 'b', 2: 'c', 3: 'd'}
print(d)        # {0: 'a', 1: 'b', 2: 'c', 3: 'd'}
print(d[0])     # 'a'
print(d[2])     # 'c'
````

In the expressions `MLB_team[1]`, `d[0]`, and `d[2]`, the numbers in square brackets appear as though they might be indices. But they have **nothing to do with the order** of items in the dictionary. **Python interprets them as dictionary keys.** If you define the same dictionary in reverse order, you still get the same values using the same keys:

````python
d = {3: 'd', 2: 'c', 1: 'b', 0: 'a'}
d        # {3: 'd', 2: 'c', 1: 'b', 0: 'a'}
d[0]     # 'a'
d[2]     # 'c'
# Why print last line code?
````

The syntax may look similar, but **you cannot treat a dictionary like a list**:

````python
type(d)   # <class 'dict'>
d[-1]     # KeyError: -1 (Negative indexing doesn't work)
d[0:2]    # Error (Slicing doesn't work on dictionaries)
````

> **Note on Dictionary Order:** Although access to items in a dictionary does not depend on order, Python **does guarantee** that the order of items in a dictionary is **preserved**. This was added as part of the Python language specification in **version 3.7** (and was true in practice in 3.6 due to implementation details, but not officially guaranteed).

### 1.4 Building a Dictionary Incrementally

Defining a dictionary using curly braces and a list of key-value pairs is fine if you know all the keys and values in advance. But what if you want to build a dictionary on the fly?

You can start by creating an **empty dictionary**, which is specified by empty curly braces. Then you can add new keys and values one at a time:

````python
person = {}
type(person)  # <class 'dict'>
# Incrementally adding key-value pairs
person['fname'] = 'Joe'
person['lname'] = 'Fonebone'
person['age'] = 51
person['spouse'] = 'Edna'
person['children'] = ['Ralph', 'Betty', 'Joey']
person['pets'] = {'dog': 'Fido', 'cat': 'Sox'}
````

Once the dictionary is created in this way, its values are accessed the same way as any other dictionary:

````python
person
# {'fname': 'Joe',
#  'lname': 'Fonebone',
#  'age': 51,
#  'spouse': 'Edna',
#  'children': ['Ralph', 'Betty', 'Joey'],
#  'pets': {'dog': 'Fido', 'cat': 'Sox'}}
print(person['fname'])       # 'Joe'
print(person['age'])         # 51
print(person['children'])    # ['Ralph', 'Betty', 'Joey']
````

**Retrieving values in sublists or subdictionaries** requires an additional index or key:

````python
print(person['children'][-1])     # 'Joey'
print(person['pets']['cat'])      # 'Sox'
print(person['pets']['dog'] == 'Fido')  # True
````

This example exhibits another feature of dictionaries: **the values contained in the dictionary don't need to be the same type**. In `person`, some values are strings, one is an integer, one is a list, and one is another dictionary.

Just as the values in a dictionary don't need to be of the same type, **the keys don't either**:

````python
foo = {42: 'aaa', 2.78: 'bbb', True: 'ccc'}
foo         # {42: 'aaa', 2.78: 'bbb', True: 'ccc'}
foo[42]     # 'aaa'
foo[2.78]   # 'bbb'
foo[True]   # 'ccc'
````

Here, one key is an integer, one is a float, and one is a Boolean.

### 1.5 Restrictions on Dictionary Keys

Almost any type of value can be used as a dictionary key in Python. You can even use built-in objects like types and functions:

````python
d = {int: 1, float: 2, bool: 3}
d           # {int: 1, float: 2, bool: 3}
d[float]    # 2
d = {bin: 1, hex: 2, oct: 3}
d[oct]      # 3
````

However, there are two restrictions that dictionary keys must abide by.

#### A) A Given Key Can Appear Only Once. Duplicate Keys Are Not Allowed.

If you assign a value to an already existing dictionary key, it does not add the key a second time, but replaces the existing value:

````python
MLB_team = {
    'Colorado' : 'Rockies',
    'Boston'   : 'Red Sox',
    'Minnesota': 'Twins',
    'Milwaukee': 'Brewers',
    'Seattle'  : 'Mariners'
}
MLB_team['Minnesota'] = 'Timberwolves'
MLB_team
# 'Minnesota': 'Timberwolves' (previous value overwritten)
# If you specify a key a second time during initial creation, the second occurrence overrides the first:
MLB_team = {
    'Colorado' : 'Rockies',
    'Boston'   : 'Red Sox',
    'Minnesota': 'Timberwolves',
    'Milwaukee': 'Brewers',
    'Seattle'  : 'Mariners',
    'Minnesota': 'Twins'  # This overrides the previous value
}
MLB_team
# 'Minnesota': 'Twins'
````

B) A Dictionary Key Must Be of an Immutable (Hashable) Type.

You have seen examples where several immutable types—integer, float, string, and Boolean—have served as dictionary keys.

**Tuples can also be dictionary keys**, because tuples are immutable:

````python
d = {(1, 1): 'a', (1, 2): 'b', (2, 1): 'c', (2, 2): 'd'}
d[(1, 1)]  # 'a'
d[(2, 1)]  # 'c'
````

> **Recall from the Tuples discussion:** One rationale for using a tuple instead of a list is that there are circumstances where an immutable type is required. This is one of them.

However, **neither a list nor another dictionary can serve as a dictionary key**, because lists and dictionaries are mutable:

````python
d = {[1, 1]: 'a', [1, 2]: 'b', [2, 1]: 'c', [2, 2]: 'd'}
# TypeError: unhashable type: 'list'
d = {'a': 1, 'b': 2}
{d: 'value'}
# TypeError: unhashable type: 'dict'
````

> **Technical Note: Why does the error message say "unhashable"?**
> 
> Technically, it is not quite correct to say an object must be immutable to be used as a dictionary key. More precisely, an object must be **hashable**, which means it can be passed to a **hash function**. A hash function takes data of arbitrary size and maps it to a relatively simpler fixed-size value called a **hash value** (or simply hash), which is used for table lookup and comparison.

````python
hash('foo')        # Output: some integer (e.g., 11132615637596761)
hash([1, 2, 3])    # TypeError: unhashable type: 'list'
````

> All of the built-in immutable types you have learned about so far are hashable, and the mutable container types (lists and dictionaries) are not. So for present purposes, you can think of hashable and immutable as more or less synonymous.

---


### 1.6 Restrictions on Dictionary Values

**There are no restrictions on dictionary values. Literally none at all.** A dictionary value can be any type of object Python supports, including mutable types like lists and dictionaries, and user-defined objects.

There is also no restriction against a particular value appearing in a dictionary multiple times:

````python
d = {0: 'a', 1: 'a', 2: 'a', 3: 'a'}
d                       # {0: 'a', 1: 'a', 2: 'a', 3: 'a'}
d[0] == d[1] == d[2]   # True
````

## Section 2: Sets

In mathematics, a set can be thought of simply as a well-defined collection of distinct objects, typically called **elements** or **members**.

Grouping objects into a set can be useful in programming as well, and Python provides a built-in `set` type to do so. Sets are distinguished from other object types by the unique operations that can be performed on them.

### Set Characteristics:

1. **Sets are unordered.**
    
2. **Set elements are unique.** Duplicate elements are not allowed.
    
3. **A set itself may be modified, but the elements contained in the set must be of an immutable type.**
    

---


### 2.1 Defining a Set

There are two ways to create a set in Python:

**Method 1: Using the `set()` Function

```text
x = set(<iter>)
```

Here, `<iter>` is an iterable (e.g., a list, tuple, or string) that generates the list of objects to be included in the set.

````python
# From a list
x = set(['foo', 'bar', 'baz', 'foo', 'qux'])
print(x)  # {'bar', 'baz', 'foo', 'qux'} (Duplicates removed)
# From a tuple
x = set(('foo', 'bar', 'baz', 'foo', 'qux'))
print(x)  # {'bar', 'baz', 'foo', 'qux'}
# From a string
s = 'quux'
list(s)   # ['q', 'u', 'u', 'x']
set(s)    # {'q', 'u', 'x'} (Unordered, duplicates removed)
````

As demonstrated:

- **The resulting sets are unordered:** The original order is not necessarily preserved.
    
- **Duplicate values are represented only once in the set.**
    

**Method 2: Using Curly Braces `{}`**

```text
x = {<obj>, <obj>, ..., <obj>}
```

When a set is defined this way, each `<obj>` becomes a distinct element of the set, even if it is an iterable itself.


````python
x = {'foo', 'bar', 'baz', 'foo', 'qux'}
print(x)  # {'bar', 'baz', 'foo', 'qux'}
x = {'q', 'u', 'u', 'x'}
print(x)  # {'q', 'u', 'x'}
````

**Summary:**

- The argument to `set()` is an iterable. It generates a list of elements to be placed into the set.
    
- The objects in curly braces are placed into the set intact, even if they are iterable.
    

**Observe the difference between these two set definitions:**

````python
{'foo'}    # {'foo'}  → A set containing the single string 'foo'
set('foo') # {'f', 'o'}  → A set of the characters in the string 'foo'
````

Empty Set:**  
Python interprets empty curly braces (`{}`) as an empty dictionary. Therefore, **the only way to define an empty set is with the `set()` function**:

````python
x = set()
type(x)  # <class 'set'>
print(x) # set()
x = {}
type(x)  # <class 'dict'>  ← Note: This is a dictionary!
````

An empty set is falsy in a Boolean context:

````python
x = set()
bool(x)      # False
x or 1       # 1
x and 1      # set()
````

**Set elements can be of different types (as long as they are immutable):**

````python
x = {42, 'foo', 3.14159, None}
print(x)  # {3.14159, 42, None, 'foo'}
# A tuple can be included in a set (because it's immutable):
x = {42, 'foo', (1, 2, 3), 3.14159}
print(x)  # {(1, 2, 3), 3.14159, 42, 'foo'}
# But lists and dictionaries cannot (because they are mutable):
a = [1, 2, 3]
# {a}  # TypeError: unhashable type: 'list'
d = {'a': 1, 'b': 2}
# {d}  # TypeError: unhashable type: 'dict'
````

### 2.2 Set Size and Membership

The `len()` function returns the number of elements in a set, and the `in` and `not in` operators can be used to test for membership:

````python
x = {'foo', 'bar', 'baz'}
len(x)          # 3
'bar' in x      # True
'qux' in x      # False
````


---

### 2.3 Set Performance Compared to Lists

Sets are **significantly faster** than lists for **membership testing**. This is because sets are implemented using a **hash table**.

````python
myset = set(range(10000))
mylist = list(range(10000))
````

````python
%timeit 9999 in myset
%timeit 9999 in mylist
````

> **Conclusion:** For large-scale membership testing, **sets** are vastly more efficient than lists.

## Section 3: Summary and Assignments 📝

### What We Learned:

1. **Dictionaries** are collections of **key-value** pairs. Elements are accessed by **key** (not index).
    
2. Dictionaries are **mutable and dynamic**. You can add, delete, or modify key-value pairs.
    
3. Dictionary **keys** must be **unique** and **immutable/hashable** (e.g., numbers, strings, tuples). Lists and dictionaries cannot be keys.
    
4. Dictionary **values** have **no restrictions**. They can be any type of object.
    
5. **Sets** are **unordered** collections of **unique** elements.
    
6. Set elements must be **immutable**. The set itself is **mutable**.
    
7. **Sets** are **much faster** than lists for membership testing (`in`) due to their hash table implementation.
    
8. An empty set can only be created with `set()` (since `{}` creates an empty dictionary).

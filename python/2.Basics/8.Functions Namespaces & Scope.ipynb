Part 1: Defining Your Own Python Functions

---

#### 1. Why Functions?

|Benefit|Description|
|---|---|
|**Abstraction & Reusability**|Write once, call many times. Change in one place. (DRY Principle)|
|**Modularity**|Break complex tasks into smaller, manageable steps.|
|**Namespace Separation**|Variables inside a function are isolated from the rest of the program.|

---


#### 2. Defining and Calling a Function

```python
def <function_name>([<parameters>]):
    """Optional docstring"""
    <statement(s)>
```

**Example:**

````python
def f():
    s = '-- Inside f()'
    print(s)
print('Before calling f()')
f()
print('After calling f()')
# Output: Before calling f()  /  -- Inside f()  /  After calling f()
````

- `def` creates the function definition (does not execute it).
    
- The function executes only when **called** (`f()`).
    
- Empty parentheses are required, even with no arguments.
    
- Use `pass` for a **stub** function (placeholder).
    

---

#### 3. Argument Passing

|Argument Type|Description|Example|
|---|---|---|
|**Positional (Required)**|Matched by order and number.|`def f(a, b, c): ...` → `f(1, 2, 3)`|
|**Keyword**|Matched by name. Order doesn't matter.|`f(c=3, a=1, b=2)`|
|**Default (Optional)**|Has a default value if omitted.|`def f(a, b=0): ...` → `f(1)`|
|**Mixed**|Positional must come **before** keyword.|`f(1, c=3, b=2)`|

**⚠️ Mutable Default Parameters (Important Pitfall!):**  
Default values are evaluated **once** when the function is defined, not each time it's called.

````python
# WRONG - List keeps growing!
def f(my_list=[]):
    my_list.append('###')
    return my_list
f()  # ['###']
f()  # ['###', '###']  ← Unexpected!
# CORRECT - Use None as sentinel
def f(my_list=None):
    if my_list is None:
        my_list = []
    my_list.append('###')
    return my_list
````


---

#### 4. Pass-by-Value vs Pass-by-Reference

Python uses **pass-by-assignment** (or pass-by-object-reference):

|Argument Type|Behavior|
|---|---|
|**Immutable** (`int`, `str`, `tuple`...)|Cannot be modified by the function. Reassignment creates a local copy.|
|**Mutable** (`list`, `dict`, `set`...)|Cannot be reassigned, but can be **modified in place**.|



````python
# Immutable - no change
def f(x):
    x = 'foo'
a = 10
f(a)
print(a)  # 10 (unchanged)
# Mutable - can modify in place
def f(x):
    x[0] = '---'
my_list = ['foo', 'bar']
f(my_list)
print(my_list)  # ['---', 'bar'] (changed!)
````

Key Rule:** A function can modify a mutable argument's **contents**, but cannot reassign the argument itself.

---

#### 5. The `return` Statement

- **Exits** the function immediately.
    
- **Passes data** back to the caller.
    
- A function without `return` (or with bare `return`) returns `None`.
    


````python
def f():
    return 'foo'
s = f()          # s = 'foo'
# Multiple return values (packed as tuple)
def f():
    return 'foo', 'bar', 'baz'
t = f()          # ('foo', 'bar', 'baz')
a, b, c = f()    # Unpacking: a='foo', b='bar', c='baz'
````

Use return values instead of side effects** whenever possible for clearer, more maintainable code.

---

#### 6. Variable-Length Argument Lists

|Syntax|Purpose|Example|
|---|---|---|
|`*args`|Packs positional arguments into a **tuple**.|`def f(*args): ...`|
|`**kwargs`|Packs keyword arguments into a **dict**.|`def f(**kwargs): ...`|


````python
def f(*args, **kwargs):
    print(args)      # Tuple of positional arguments
    print(kwargs)    # Dict of keyword arguments
f(1, 2, 'foo', x=100, y=200)
# Output: (1, 2, 'foo')   {'x': 100, 'y': 200}
````

**Unpacking in function calls:**

````python
def f(x, y, z):
    print(x, y, z)
a = [1, 2, 3]
f(*a)              # Unpacks list → f(1, 2, 3)
d = {'x': 1, 'y': 2, 'z': 3}
f(**d)             # Unpacks dict → f(x=1, y=2, z=3)
````

**All three can be combined (must be in this order):**

```python
def f(a, b, *args, **kwargs):
    ...

```


---

#### 7. Keyword-Only and Positional-Only Arguments

|Feature|Syntax|Description|
|---|---|---|
|**Keyword-Only**|`def f(*, kw):` or `def f(*args, kw):`|Arguments after `*` must be passed by keyword.|
|**Positional-Only** (Python 3.8+)|`def f(a, b, /, c):`|Arguments before `/` must be positional.|


````python
# Keyword-only
def concat(*args, prefix='-> ', sep='.'):
    print(f'{prefix}{sep.join(args)}')
concat('a', 'b', 'c')                    # -> a.b.c
concat('a', 'b', 'c', prefix='//', sep='-')  # //a-b-c
# Positional-only (Python 3.8+)
def f(x, y, /, z):
    print(x, y, z)
f(1, 2, 3)        # OK
f(x=1, y=2, z=3)  # TypeError! x and y are positional-only
````

#### 8. Docstrings

```python
def avg(*args):
    """Returns the average of a list of numeric values."""
    return sum(args) / len(args)
# Multi-line docstring
def foo(bar=0, baz=1):
    """Perform a foo transformation.
    Keyword arguments:
    bar -- magnitude along the bar axis (default=0)
    baz -- magnitude along the baz axis (default=1)
    """
    <function_body>
    Access via `function.__doc__` or `help(function)`.


---

#### 9. Function Annotations


````python
def f(a: int, b: str = 'baz') -> float:
    print(a, b)
    return 3.5
f.__annotations__  # {'a': int, 'b': str, 'return': float}
````


- Annotations provide **metadata** (types, descriptions, etc.).
    
- They are **not enforced** by Python (purely documentation).
    
- Stored in `function.__annotations__` dictionary.

### Part 2: Namespaces and Scope

---

#### 10. Four Types of Namespaces

|Namespace|Description|Lifetime|
|---|---|---|
|**Built-in**|Names of all Python built-in objects (`print`, `len`, `int`...)|Created at startup, exists until interpreter exits|
|**Global**|Names defined at the main program level|Created when main program starts|
|**Enclosing**|Names in the outer function (for nested functions)|Duration of outer function execution|
|**Local**|Names defined inside a function|Duration of function execution|

A **namespace** is like a dictionary mapping names to objects.

---

#### 11. The LEGB Rule (Scope Resolution)

Python searches for a name in this order:

**L**ocal → **E**nclosing → **G**lobal → **B**uilt-in

````python
x = 'global'       # Global scope
def f():
    x = 'enclosing'  # Enclosing scope
    
    def g():
        x = 'local'    # Local scope
        print(x)       # Prints: 'local' (found in local first)
    
    g()
f()
````


|Example|Where `x` is defined|Output|
|---|---|---|
|Only global|Global|`'global'`|
|Global + Enclosing|Enclosing wins|`'enclosing'`|
|Global + Enclosing + Local|Local wins|`'local'`|
|Nowhere|-|`NameError`|

---

#### 12. `globals()` and `locals()

````python
x = 'foo'
globals()         # Dictionary of global namespace
globals()['x']    # 'foo' (access via dictionary)
def f(x, y):
    s = 'foo'
    print(locals())  # {'x': ..., 'y': ..., 's': 'foo'}
````


|Function|Returns|
|---|---|
|`globals()`|Reference to global namespace dict (modifiable)|
|`locals()`|**Copy** of local namespace dict (outside function: same as `globals()`)|

---

#### 13. Modifying Variables Across Scopes

|Keyword|Purpose|
|---|---|
|`global`|Access and modify a **global** variable from inside a function.|
|`nonlocal`|Access and modify a variable in the **nearest enclosing scope** (not global).|



````python
# global
x = 20
def f():
    global x
    x = 40     # Modifies global x
f()
print(x)        # 40
# nonlocal
def f():
    x = 20
    def g():
        nonlocal x
        x = 40  # Modifies x in enclosing scope (f)
    g()
    print(x)    # 40
````

⚠️ Important Rule:** If you assign to a variable inside a function, Python treats it as **local** unless declared `global` or `nonlocal`.

````python
x = 4
def f():
    print(x)   # ERROR! (x is local because it's assigned below, but used before assignment)
    x = 3
````


**Best Practice:** Avoid `global` and `nonlocal` when possible. Use function **return values** instead.

---

#### 14. Summary

|Concept|Key Points|
|---|---|
|`def`|Creates a function.|
|Positional args|Matched by order. Must come before keyword args.|
|Keyword args|Matched by name. Any order.|
|Default args|`def f(a=10):` – Use `None` for mutable defaults.|
|`*args` / `**kwargs`|Variable-length positional/keyword arguments.|
|`return`|Exits function, passes data back.|
|Pass-by-assignment|Immutable args can't be changed; mutable args can be modified in place.|
|Docstrings|`""" ... """` – Access via `__doc__` or `help()`.|
|Annotations|`def f(a: int) -> str:` – Metadata, not enforced.|
|Namespace|Mapping of names to objects.|
|LEGB|Local → Enclosing → Global → Built-in.|
|`global`|Modify global variable from function.|
|`nonlocal`|Modify enclosing scope variable from nested function.|

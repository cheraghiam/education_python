### 1. The `if` Statement

The `if` statement enables **conditional execution** based on a condition.

```python
if <expr>:
    <statement>

- `<expr>` is evaluated in a Boolean context (truthy/falsy).
    
- The colon (`:`) after `<expr>` is **required**.
    
- If `<expr>` is **truthy** → the block executes.
    
- If `<expr>` is **falsy** → the block is skipped.
    
```

### Falsy Values in Python

In Python, the following values are considered **falsy** (evaluate to `False` in a Boolean context):

|Falsy Value|Description|
|---|---|
|`False`|The Boolean false value|
|`None`|The null/absent value|
|`0`|Integer zero|
|`0.0`|Floating-point zero|
|`0j`|Complex zero|
|`""`|Empty string|
|`''`|Empty string (single quotes)|
|`[]`|Empty list|
|`()`|Empty tuple|
|`{}`|Empty dictionary|
|`set()`|Empty set|
|`range(0)`|Empty range|

**Everything else** is truthy.

````python
# Falsy examples
bool(False)    # → False
bool(None)     # → False
bool(0)        # → False
bool(0.0)      # → False
bool("")       # → False
bool([])       # → False
bool({})       # → False
bool(set())    # → False
# Truthy examples
bool(True)     # → True
bool(1)        # → True
bool(-5)       # → True
bool(3.14)     # → True
bool("Hello")  # → True
bool(" ")      # → True (space is a character!)
bool([0])      # → True (non-empty list)
bool([[]])     # → True (non-empty list containing an empty list)
````

Important:** `" "` (a string with a space) is **truthy**. `[0]` (a list containing zero) is also **truthy** because the container is non-empty—only empty containers are falsy.

````python
x, y = 0, 5
if x < y:        # True → executes
    print('yes')  # Output: yes
if x:            # False (0 is falsy) → skipped
    print('yes')  # (no output)
if x or y:       # True (y is truthy) → executes
    print('yes')  # Output: yes
if 'aul' in 'grault':  # True → executes
    print('yes')  # Output: yes
````

### 2. Indentation and Blocks

Python uses **indentation** (the off-side rule) to define **blocks** (suites). All statements at the same indentation level belong to the same block. The end of a block is indicated by a line with **less** indentation.

```python
if <expr>:
    <statement_1>
    <statement_2>
    <statement_3>
<following_statement>   # Less indented → block ends
```

**Nested blocks:**

````python
if 'foo' in ['foo', 'bar']:    # True
    print('Outer is true')
    
    if 10 > 20:                # False
        print('Inner 1')       # Skipped
    
    print('Between inners')
    
    if 10 < 20:                # True
        print('Inner 2')       # Executes
    print('End of outer')
print('After outer')
````

### 3. The `else` and `elif` Clauses

#### `else` – Execute when the condition is false:

````python
x = 120
if x < 50:
    print('x is small')
else:
    print('x is large')
# Output: x is large
````

#### `elif` – Multiple branches with **short-circuit** evaluation (stops at the first truthy condition):

````python
name = 'Joe'
if name == 'Fred':
    print('Hello Fred')
elif name == 'Xander':
    print('Hello Xander')
elif name == 'Joe':
    print('Hello Joe')
elif name == 'Arnold':
    print('Hello Arnold')
else:
    print("I don't know who you are!")
# Output: Hello Joe
````

Important:** Once a condition is true, remaining conditions are **not evaluated**:

````python
if 'a' in 'bar':
    print('foo')
elif 1/0:                    # Never evaluated! No error.
    print("Won't happen")
else:
    print("Won't happen either")
# Output: foo
````

**Pythonic alternative** to long `elif` chains:

````python
name = 'Joe'
if name in ['Fred', 'Xander', 'Joe', 'Arnold']:
    print(f'Hello {name}')
else:
    print("I don't know who you are!")
````

### 4. One-Line `if` Statements

Python allows writing `if` statements on one line, but this is **discouraged** (PEP 8) except for very simple cases.

````python
if 'f' in 'foo': print('1'); print('2'); print('3')
# Output: 1 / 2 / 3
if 'z' in 'foo': print('1'); print('2'); print('3')
# (no output)

#Acceptable use for debugging:

debugging = True
if debugging: print('About to call function foo()')
````

### 5. Conditional Expressions (Ternary Operator)

A compact way to embed a condition inside an **expression**:

```python
<expr1> if <conditional_expr> else <expr2>

- If `<conditional_expr>` is true → returns `<expr1>`.
    
- If `<conditional_expr>` is false → returns `<expr2>`.
    
```

````python
age = 12
status = 'minor' if age < 21 else 'adult'
print(status)  # minor
# With boolean:
raining = True
print("Let's go to the", 'beach' if not raining else 'library')
# Output: Let's go to the library
````

**Replaces simple if/else assignments:**

````python

# Instead of:
a = 5
b = 10
if a > b:
    m = a
else:
    m = b
# Use:
m = a if a > b else b
m
````

**Short-circuit evaluation:**

````python
'foo' if True else 1/0   # Returns 'foo' (1/0 never evaluated, no error)
1/0 if False else 'bar'  # Returns 'bar' (1/0 never evaluated, no error)
````

### 6. The `pass` Statement

A **placeholder** that does nothing. Useful for code stubs where syntax requires a statement but you haven't implemented the logic yet.

````python
if True:
    pass    # No error – syntax requirement satisfied
print('foo')  # Output: foo




# Practical use:

name = 'Ali'
if name == 'Ali':
    # TODO: add name to database
    pass
else:
    # TODO: warn user about registration
    pass
````


> **Without `pass`**, an empty block would raise an `IndentationError`.

### 7. Summary

|Concept|Key Points|
|---|---|
|Falsy Values|`False`, `None`, `0`, `0.0`, `""`, `[]`, `()`, `{}`, `set()`, `range(0)`|
|Truthy Values|Everything else (non-zero numbers, non-empty strings/containers, etc.)|
|`if`|Executes a block only if the condition is truthy.|
|Indentation|Defines blocks (suites) in Python. Same indentation = same block.|
|`else`|Executes when the `if` condition is false.|
|`elif`|Multiple conditions, short-circuit evaluation. Stops at first truthy.|
|One-Line `if`|Possible but discouraged for complex code.|
|Conditional Expression|`x if cond else y` – returns a value. Use for simple branching inside expressions.|
|`pass`|Placeholder that does nothing. Satisfies syntax requirement for empty blocks.|

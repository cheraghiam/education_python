### Part 1: List Comprehensions

---

#### 1. What Are List Comprehensions?

List comprehensions are a concise, Pythonic way to create lists in a single line of code. They are often described as more readable and declarative than traditional `for` loops.

---

#### 2. Three Ways to Create Lists

**Method 1: Traditional `for` Loop (3 steps)

````python
squares = []
for i in range(10):
    squares.append(i * i)
````

- Create an empty list.
    
- Loop over an iterable.
    
- Append each element.
    

**Method 2: `map()` Function (Functional Approach)

````python
txns = [1.09, 23.56, 57.84, 4.56, 6.78]
TAX_RATE = .08
def get_price_with_tax(txn):
    return txn * (1 + TAX_RATE)
final_prices = map(get_price_with_tax, txns)
list(final_prices)  # Returns a list from the map object
````

- Pass a function and an iterable to `map()`.
    
- Returns a `map` object (not a list—must be converted).
    

**Method 3: List Comprehension (Pythonic Way)

````python
squares = [i * i for i in range(10)]
# Result: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
````

**Basic Syntax:**

```text
new_list = [expression for member in iterable]
```

Three elements:

- **expression:** The value to include in the new list (e.g., `i * i`).
    
- **member:** The loop variable (e.g., `i`).
    
- **iterable:** The source of values (e.g., `range(10)`).
    

---

#### 3. Benefits of List Comprehensions

|Benefit|Description|
|---|---|
|**Single Tool**|Handles creation, mapping, and filtering—no need for separate approaches.|
|**Pythonic**|Embraces Python's philosophy: simple, powerful tools for many situations.|
|**Declarative**|Focus on **what** you want, not **how** to build it.|
|**Performance**|Often faster than equivalent `for` loops.|

---

#### 4. Adding Conditional Logic

**Filtering (condition at the end):**  
Includes only items that satisfy the condition.

```python
new_list = [expression for member in iterable (if conditional)]
```

````python
sentence = 'the rocket came back from mars'
vowels = [i for i in sentence if i in 'aeiou']
# Result: ['e', 'o', 'e', 'a', 'e', 'a', 'o', 'a']
````

**Complex filter using a function:**

````python
def is_consonant(letter):
    vowels = 'aeiou'
    return letter.isalpha() and letter.lower() not in vowels
consonants = [i for i in sentence if is_consonant(i)]
````

Modifying values (condition in the expression):**  
Changes the value based on a condition.

```text
new_list = [expression (if conditional) for member in iterable]
```

````python
original_prices = [1.25, -9.45, 10.22, 3.78, -5.92, 1.16]
prices = [i if i > 0 else 0 for i in original_prices]
# Result: ?
````

**Summary of condition placement:**

| Placement                 | Purpose                      | Example                          |
| ------------------------- | ---------------------------- | -------------------------------- |
| End (after `for`)         | **Filter** – keep or discard | `[i for i in x if i > 0]`        |
| Beginning (in expression) | **Transform** – change value | `[i if i > 0 else 0 for i in x]` |

---

#### 5. Set and Dictionary Comprehensions

**Set Comprehension** (no duplicates, curly braces `{}`):

````python
quote = "life, uh, finds a way"
unique_vowels = {i for i in quote if i in 'aeiou'}
# Result: {'a', 'e', 'i', 'u'}
````

**Dictionary Comprehension** (key-value pairs):

````python
squares = {i: i * i for i in range(10)}
# Result: {0: 0, 1: 1, 2: 4, 3: 9, ..., 9: 81}
````

**Comparison of comprehensions:**

|Type|Brackets|Output|
|---|---|---|
|List|`[ ]`|List|
|Set|`{ }` (no colon)|Set (unique, unordered)|
|Dictionary|`{k: v}`|Dictionary (key-value pairs)|

---

#### 6. Nested Comprehensions

**Creating Matrices:**

````python
matrix = [[i for i in range(5)] for _ in range(6)]
# Creates 6 rows, each with values [0, 1, 2, 3, 4]
````

**Dictionary with nested lists:**

````python
cities = ['Austin', 'Tacoma', 'Topeka', 'Sacramento', 'Charlotte']
temps = {city: [0 for _ in range(7)] for city in cities}
````

**Flattening a matrix:**

````python
flat = [num for row in matrix for num in row]
````


---

#### 7. When NOT to Use List Comprehensions

|Issue|Warning|
|---|---|
|**Nested Comprehensions**|Can become hard to read. If logic is complex, use a `for` loop instead.|
|**Large Data**|List comprehensions build the entire list in memory. Consider generators for huge datasets.|
|**Side Effects**|If the expression has side effects (e.g., printing), avoid comprehensions.|

**Example: When a loop is clearer than a comprehension:**


````python
# Hard to read nested comprehension
flat = [num for row in matrix for num in row]
# Clearer as a for loop
flat = []
for row in matrix:
    for num in row:
        flat.append(num)
````

> **Golden Rule:** Write code your team can easily understand and modify. If a comprehension hurts readability, use a loop.

### Part 2: Basic Input, Output, and String Formatting

---

#### 8. Reading Input with `input()

```text
input([<prompt>])
```

- Pauses execution and waits for the user to type a line.
    
- Returns the input as a **string**.
    
- The newline (`Enter`) is **not** included in the returned string.
    


````python
s = input()                  # User types: some text
print(s)                     # 'some text'
name = input('What is your name? ')  # Displays prompt before input
print(name)                  # Whatever the user typed
````

**⚠️ Important: `input()` always returns a string. Convert to numeric types when needed:**

````python
n = input('Enter a number: ')   # User types: 10
print(n + 100)                   # TypeError! (string + int)
n = int(input('Enter a number: '))  # Convert to int
print(n + 100)                      # 110 (works!)
````


---

#### 9. Output with `print()

```text
print(<obj>, ..., <obj>)
```

- Displays string representations of objects to the console.
    
- By default: separates objects by a **single space**, appends a **newline** at the end.
    

````python
fname = 'Winston'
lname = 'Smith'
print('Name:', fname, lname)
# Output: Name: Winston Smith
# Works with any type
a = [1, 2, 3]
b = -12
d = {'foo': 1, 'bar': 2}
print(a, b, d)
# Output: [1, 2, 3] -12 {'foo': 1, 'bar': 2}
````


---

#### 10. The `sep=` Keyword Argument

Controls the **separator** between objects (default: `' '`).

````python
print('foo', 42, 'bar')
# Output: foo 42 bar
print('foo', 42, 'bar', sep='/')
# Output: foo/42/bar
print('foo', 42, 'bar', sep='...')
# Output: foo...42...bar
print('foo', 42, 'bar', sep='')
# Output: foo42bar
# Useful for formatting output
d = {'foo': 1, 'bar': 2, 'baz': 3}
for k, v in d.items():
    print(k, v, sep=' -> ')
# Output:
# foo -> 1
# bar -> 2
# baz -> 3
````


---

#### 11. The `end=` Keyword Argument

Controls the **string appended at the end** of output (default: `'\n'`).

````python
# Default behavior (newline at end)
for n in range(3):
    print(n)
# Output:
# 0
# 1
# 2
# Custom end (no newline)
for n in range(3):
    print(n, end=' ')
# Output: 0 1 2
# Using end to chain print statements
print('foo', end='/')
print(42, end='/')
print('bar')
# Output: foo/42/bar
````


---

#### 12. Output Stream Keyword Arguments (Advanced)

|Argument|Description|
|---|---|
|`file=<stream>`|Sends output to a file instead of the console.|
|`flush=True`|Forces immediate writing to the output stream.|

These will be covered in detail in the File I/O section later.

---

#### 13. Summary

|Concept|Key Points|
|---|---|
|List Comprehension|`[expr for member in iterable]` – concise list creation.|
|Filtering|`[expr for member in iterable if condition]`|
|Transformation|`[expr if condition else alt for member in iterable]`|
|Set Comprehension|`{expr for member in iterable}` – unique values.|
|Dictionary Comprehension|`{key: value for member in iterable}`|
|Nested Comprehensions|Powerful but can reduce readability.|
|`input()`|Reads user input, always returns a string.|
|`print()`|Displays output to console.|
|`sep=`|Separator between objects (default: space).|
|`end=`|String appended at end (default: newline).|

## Section 1: Ways to Interact with Python 🐍

Before learning to code, we must understand how to "talk" to Python. Fortunately, there are several ways to do this.

### 1.1 Interactive Mode: REPL and IPython

The simplest way to interact with Python is using the **REPL** (Read-Evaluate-Print Loop). The acronym stands for the four steps the interpreter performs:

- **R**ead: Reads the command you enter.
    
- **E**valuate: Evaluates and executes the command.
    
- **P**rint: Prints the output (if any) to the console.
    
- **L**oop: Loops back and repeats the process.
    

#### 1.1.1 Starting the Standard Python Interpreter

Open **Command Prompt** on Windows or **Terminal** on macOS/Linux and type:


```bash
python
# Or if you also have Python 2 installed:
python3
```
The output should look similar to this (your version may differ):

```text
Python 3.9.25 (main, Apr  8 2025, 10:30:00) [Clang 14.0.0 ] on darwin
Type "help", "copyright", "credits" or "license" for more information.
>>>
```

The `>>>` prompt indicates that the interpreter is ready to receive your commands. Now you can execute your first line of code:

```python
>>> print("Hello, World")
Hello, World
```

**Useful Tip:** In a REPL environment, you don't even need to use `print()` to display a value. Just type the name of the variable or object and press Enter:

```python
>>> 0x10
16
>>> "Salam"
'Salam'
```
#### 1.1.2 Exiting the Interpreter

There are three ways to exit a REPL session:

1. Type `exit()` and press Enter.
    
2. On Linux/macOS: Press `Ctrl+D`.
    
3. On Windows: Press `Ctrl+Z` then Enter.
    

#### 1.1.3 IPython (An Enhanced REPL)

[IPython](https://ipython.org/) is an upgraded version of the standard REPL that adds the following features:

- Line numbering (`In [1]`, `Out[1]`)
    
- Syntax highlighting with colors
    
- Prettier display of arrays and dataframes
    
- Smarter tab-completion
    

**Installation:**

```bash
pip install ipython
```

**Sample IPython Session:**


```python

In [2]: digits = np.array([
   ...:     [1, 2, 3],

   ...:     [6, 7, 9],
```

### 1.2 Running a Python Script from the Command Line

For larger projects, typing code directly into the REPL becomes inefficient. Instead, we write code in files with a `.py` extension.

**Steps:**

1. Create a new text file named `hello.py`.
    
2. Write the following code inside it:

    ```python
    print("Hello, World!")
	```

3. Save the file.
    
4. Open a terminal and navigate to the file's location.
    
5. Run the following command:
    

```bash
python hello.py
# Output: Hello, World!
```

If the script is located elsewhere, you can specify the full path:

```bash
python /home/username/Documents/test/hello.py
```

> **Important Note:** A script file is not strictly _required_ to have a `.py` extension. However, using this extension is a useful convention that makes Python files easily identifiable and allows your operating system to associate them with the correct application.

Section 2: Basic Data Types in Python 🧮

In Python, **everything is an object**. Let's explore the most important built-in data types.

### 2.1 Integers (`int`)

In Python 3, integers have **no fixed length limit** (bounded only by your system's available memory).

````python
# Simple definition
x = 300
print(x)        # 300
# Very large numbers work fine
big_num = 123123123123123123123123123123123123123123123123
print(big_num)  # Prints the exact same number
# Checking the type
type(10)        # <class 'int'>
````

2.1.1 Representing Integers in Different Bases

Python allows you to define integers in binary (base 2), octal (base 8), and hexadecimal (base 16) notations:

|Prefix|Base|Example|Decimal Value|
|---|---|---|---|
|`0b` / `0B`|Binary (2)|`0b10`|2|
|`0o` / `0O`|Octal (8)|`0o10`|8|
|`0x` / `0X`|Hexadecimal (16)|`0x10`|16|


````python
print(0b10)     # 2 (binary)
print(0o10)     # 8 (octal)
print(0x10)     # 16 (hex)
print(0x43)     # 67 (4 × 16 + 3 = 67)
````

The underlying type of all these values is still `int`:

````python
type(0b101010)  # <class 'int'>
````


---

### 2.2 Floating-Point Numbers (`float`)

Floats are specified using a decimal point (`.`). You can also use **Scientific Notation** with `e` or `E`.


````python
# Standard definition
pi = 3.14
half = .5          # Equivalent to 0.5
four = 4.          # Equivalent to 4.0
# Scientific notation
c = 3e8           # 300,000,000.0 (speed of light)
tiny = 4.2e-4     # 0.00042
# Checking the type
type(4.2)          # <class 'float'>
````

#### 2.2.1 Limitations of Floating-Point Representation (Very Important!)

Computers store numbers in **binary (base 2)** format. Many decimal fractions cannot be represented exactly as binary fractions (just like we can't write 1/3 exactly as a finite decimal: 0.33333...).

**The Famous Example:**

````python
0.1 + 0.2          # Output: 0.30000000000000004
0.1 + 0.2 == 0.3   # Output: False !!!
````


> **Range of Representable Values (IEEE 754 Standard):**
> 
> - **Maximum value:** approximately `1.8 × 10³⁰⁸`
>     
> - **Smallest non-zero value:** approximately `5.0 × 10⁻³²⁴`
>     

````python
1.8e308   # 1.8e+308
1.9e308   # inf (infinity)
5e-324    # 5e-324
1e-325    # 0.0 (becomes zero)
````

#### 2.2.2 How to Compare Floating-Point Numbers Correctly

To compare two floats, instead of `==`, use a **tolerance value (epsilon)**:


````python
# Correct way to compare floats
epsilon = 1e-9
abs((0.1 + 0.2) - 0.3) < epsilon   # Output: True
````

### 2.3 Complex Numbers (`complex`)

Python has built-in support for complex numbers. The format is: `<real part> + <imaginary part>j

````python
z = 2 + 3j
print(z)          # (2+3j)
type(z)           # complex
# Accessing real and imaginary parts
z.real            # 2.0
z.imag            # 3.0
````

### 2.4 Strings (`str`)

Strings are sequences of character data. They can be delimited using either **single quotes (`'`)** or **double quotes (`"`)** .

````python
s1 = "I am a string."
s2 = 'I am too.'
print(s1)         # I am a string.
type(s2)          # str
# Empty string
empty = ""
````

#### 2.4.1 The Problem of Including Quotes Inside Strings

If you need to include a quote character _as part of_ the string itself, you have two options:

**Option 1: Use a Different Quote Type for the Delimiter**


````python
#Use double quotes when the string contains a single quote
print("This string contains a single quote (') character.")
# Use single quotes when the string contains a double quote
print('This string contains a double quote (") character.')
````

**Option 2: Use the Escape Character (`\`)**

````python
print('This string contains a single quote (\') character.')
print("This string contains a double quote (\") character.")
````

2.4.2 Escape Sequences

A backslash (`\`) inside a string indicates that the character(s) following it should be treated specially.

|Sequence|Escaped Interpretation|Description|
|---|---|---|
|`\'`|`'`|Literal single quote|
|`\"`|`"`|Literal double quote|
|`\\`|`\`|Literal backslash|
|`\n`|Linefeed|New line|
|`\t`|Tab|Horizontal tab|
|`\r`|Carriage Return|Return to start of line|
|`\b`|Backspace|Backspace character|
|`\uXXXX`|Unicode 16-bit|Unicode character with 4-digit hex code|
|`\UXXXXXXXX`|Unicode 32-bit|Unicode character with 8-digit hex code|
|`\N{name}`|Unicode by name|Unicode character by its name|
|`\ooo`|Octal value|Character with octal value `ooo`|
|`\xhh`|Hex value|Character with hex value `hh`|

**Examples:**



````python
print("a\tb")           # a    b
print("a\nb")           # a (new line) b
print("\u2192")         # → (rightwards arrow)
print("\N{rightwards arrow}")  # → (same arrow)
print("\u0033")         # 3
print("a\141\x61")      # aaa
````

**Continuing a String Across Multiple Lines with `\`:**

````python
# This method *does not* add newline characters; it just allows the code to wrap.
print('a very very very \
very very very \
very long string.')
# Output: a very very very very very very very long string.
````

#### 2.4.3 Raw Strings

A raw string literal is preceded by `r` or `R`. Escape sequences in the string are **not translated**. This is extremely useful for **Windows file paths** and **Regular Expressions (Regex)** .

````python
#Normal string: \n becomes a newline
print('foo\nbar')
# foo
# bar
# Raw string: \n remains the literal characters \ and n
print(r'foo\nbar')
# foo\nbar
# Comparison for file paths
path_normal = "C:\\Users\\Name\\new_folder"   # Must double the backslashes
path_raw = r"C:\Users\Name\new_folder"        # Simple and readable
````

#### 2.4.4 Triple-Quoted Strings

Delimited by three single quotes (`'''`) or three double quotes (`"""`), these allow you to:

- Create strings that span multiple lines (no need for `\n`).
    
- Freely include both single and double quotes without escaping.
    


````python
#Multiline string
poem = """This is line 1.
This is line 2.
This is line 3."""
print(poem)
# This is line 1.
# This is line 2.
# This is line 3.
# Mixing quotes without escaping
text = '''This string has a single (') and a double (") quote.'''
````

### 2.5 Boolean Type (`bool`)

Objects of Boolean type can have one of two values: `True` or `False`. (Note the capital **T** and **F**.)


````python
type(True)          # bool
type(False)         # bool
is_student = True
has_job = False
````

#### 2.5.1 The Concept of "Truthiness" and "Falsiness"

In Python, non-Boolean objects can also be evaluated in a **Boolean context** (e.g., in an `if` statement). This is referred to as an object being "truthy" or "falsy."

**Falsy Objects (Evaluate to `False`):**

- `False`
    
- `None`
    
- Zero of any numeric type: `0`, `0.0`, `0j`
    
- Empty string: `""`
    
- Empty containers: `[]`, `()`, `{}`, `set()`, `range(0)`
    

**Truthy Objects (Evaluate to `True`):**

- Everything else!
    


````python
# Example from upcoming topics (if statements)
name = ""
if name:
    print("Name exists")   # This line will NOT execute
else:
    print("Name is empty") # This line WILL execute
````

## Section 4: Variables in Python 📛

In Python, a variable is a **symbolic name** that is a **reference** or pointer to an object in memory. This is fundamentally different from the "box containing a value" concept in languages like C.

### 4.1 Variable Assignment

In Python, you do not need to **declare** a variable before using it. You create it by assigning a value with the equals sign (`=`):

````python
n = 300
print(n)        # 300
n               # 300 (In REPL, it is displayed without print)
````

**Chained Assignment:**

````python
a = b = c = 300
print(a, b, c)  # 300 300 300
````

4.2 Variable Types in Python (Dynamic Typing)

Unlike statically-typed languages (like Java or C++), a Python variable can be reassigned to an object of a **completely different type** over its lifetime:

````python
var = 23.5
print(type(var))   # <class 'float'>
var = "Now I'm a string"
print(type(var))   # <class 'str'>
````

### 4.3 Object References

This concept is **the most important** part of understanding Python variables. Let's walk through it step-by-step:

**Step 1:** Initial assignment

````python
n = 300
````

Python creates an `int` object with the value `300`, and `n` points to it.

**Step 2:** Assigning one variable to another

````python
m = n
````

This does **not** create a new object. It simply creates a new reference, `m`, which points to the **same** `300` object.

**Step 3:** Changing the value of the second variable

````python
m = 400
````

Python now creates a _new_ `int` object with the value `400`, and `m` points to this new object. Variable `n` **still** points to the original `300` object.

**Step 4:** Changing the value of the first variable

````python
n = "foo"
````

Python creates a `str` object with the value `"foo"`, and `n` now points to it.

**What happened to the `300` object?**  
If no other variables reference the `300` object, it becomes **orphaned** and is no longer accessible. Python's **Garbage Collector** will eventually notice this and reclaim the memory. This process is automatic.

### 4.4 Object Identity with `id()`

Every object in Python has a **unique integer identifier** that remains constant for the object's lifetime. The `id()` function returns this identifier (typically the memory address).

````python
n = 300
m = n
id(n)               # e.g., 140238739082320
id(m)               # The same number as above
m is n              # True (both point to the same object)
m = 400
id(m)               # New ID, e.g., 139989024261008
id(n)               # Old ID is still for n
````

4.5 Deep Dive: Caching of Small Integer Values

For performance optimization, the Python interpreter pre-creates objects for integers in the range **[-5, 256]** at startup and reuses them. Therefore:

````python
# With small integers (in the range [-5, 256])
a = 30
b = 30
id(a) == id(b)      # True ! (Both refer to the same shared object)
# With large integers (outside the range)
x = 300
y = 300
id(x) == id(y)      # False ! (Two separate objects are created)
````

> **Moral of the story:** Use `==` to compare **values** for equality. Use `is` only to check if two names refer to the **exact same object** (commonly used for `None`).

### 4.6 Variable Naming Rules

**Official Rules:**

- Can contain uppercase and lowercase letters (`a-z`, `A-Z`), digits (`0-9`), and underscores (`_`).
    
- **Cannot** begin with a digit.
    
- Are **case-sensitive** (`age` and `Age` are different variables).
    

**Valid Examples:**

````python
name = "Bob"
Age = 54
has_W2 = True
_private = "secret"
var_1 = 100
````

**Invalid Examples:**

````python
1099_filed = False  # SyntaxError: Cannot start with a digit
my-var = 10         # SyntaxError: Hyphens are not allowed
my var = 10         # SyntaxError: Spaces are not allowed
````

#### 4.6.1 Naming Conventions (PEP 8)

PEP 8 is the official style guide for Python code. It recommends the following conventions for variable and function names:

| Style               | Description                                        | Example         | Recommended Use                        |
| ------------------- | -------------------------------------------------- | --------------- | -------------------------------------- |
| **Snake Case**      | Words separated by underscores, all lowercase      | `student_count` | **Functions and variables** (Official) |
| **Camel Case**      | First word lowercase, subsequent words capitalized | `studentCount`  | Informal/Personal preference           |
| **Pascal Case**     | All words capitalized                              | `StudentCount`  | **Class names** (Official)             |
| **Screaming Snake** | All uppercase + underscores                        | `MAX_SIZE`      | Constants                              |

**Instructor's Advice:** Train yourself from the beginning to use **Snake Case** for variables. It is the standard in the Python community.

````python
#✅ Recommended
number_of_students = 25
first_name = "Ali"
# ⚠️ Works, but is non-standard
numberOfStudents = 25
FirstName = "Ali"
````
